```python
import os 
import shutil

src= "/content/drive/MyDrive/DeepFakeProjesi/Covidli/Chunks_srt"
file= open('AllText5.txt', 'w', encoding = "ISO-8859-1") 
for file_name in os.listdir(src):
   print(file_name)
   with open(f"{src}/{file_name}", "r", encoding = "ISO-8859-1") as f:
       file_n = file_name.split(".")[0]
       file.write(file_n+".wav|"+f.read())
       file.write("\n")
```
